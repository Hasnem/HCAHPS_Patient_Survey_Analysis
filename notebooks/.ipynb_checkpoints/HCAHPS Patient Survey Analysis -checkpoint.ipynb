## 1. Executive Summary

Between **2014 and 2023**, U.S. hospitals reported patient satisfaction (HCAHPS) scores that generally hovered around **70–72%** (Top-box). Starting in **2021**, a dip of roughly 2–3% emerged, correlating with COVID-19 disruptions, followed by mild recovery in 2023. Advanced modeling (Lasso, Ridge, Elastic Net) reveals **Care Transition**, **Quietness of Hospital Environment**, and **Cleanliness** as top drivers of Overall Hospital Rating, while **Communication about Medicines** lags. 

By focusing on care transitions, environment, and clearer medication communication, hospitals can bolster patient experiences—validated by our regression models’ high predictive power (R² > 0.90).

**Observations:**

- No duplicates).

<a id="44-covid-19-impact-analysis"></a>
### 4.4 COVID-19 Impact Analysis


```python
# Change in Top-box Percentage Before vs After COVID by Measure

# Define year split
year_split = 2021

# Calculate averages for each measure before and after the COVID-19 onset
before_covid = hcahps_data[hcahps_data['Year'] < year_split].groupby('Measure')['Top-box Percentage'].mean().reset_index()
after_covid = hcahps_data[hcahps_data['Year'] >= year_split].groupby('Measure')['Top-box Percentage'].mean().reset_index()

# Merge and calculate the difference
before_after_covid = before_covid.merge(after_covid, on='Measure', suffixes=('_before', '_after'))
before_after_covid['Difference'] = before_after_covid['Top-box Percentage_after'] - before_after_covid['Top-box Percentage_before']
before_after_covid = before_after_covid.sort_values('Difference', ascending=False)

# Plot the first graph
colors = ['#1f77b4' if diff >= 0 else '#d62728' for diff in before_after_covid['Difference']]
fig, ax = plt.subplots(figsize=(14, 10), dpi=100)

ax.barh(before_after_covid['Measure'], before_after_covid['Difference'], color=colors, edgecolor='black')

# Add average change line and label
total_avg_change = before_after_covid['Difference'].mean()
ax.axvline(total_avg_change, color='blue', linestyle='--', linewidth=1)
ax.text(-2.2, 0.25, f'Avg Change ({total_avg_change:.2f})', color='blue', fontsize=12, ha='left')

# Customize plot
ax.set_title('Change in Top-box Percentage Before vs After COVID by Measure', fontsize=16, weight='bold')
ax.set_xlabel('Difference in Score', fontsize=14)
ax.set_ylabel('Measure', fontsize=14)
ax.grid(True, axis='x', linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Add text labels for each bar
for i, (diff, measure) in enumerate(zip(before_after_covid['Difference'], before_after_covid['Measure'])):
    ax.text(diff, i, f'{diff:.2f}', ha='right' if diff < 0 else 'left', va='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

- All measures saw a decline in top-box scores post-COVID, with an average decrease of **-1.72%**.  
- **Communication about Medicines** and **Responsiveness of Hospital Staff** experienced the largest drops, exceeding **-3%**, likely due to pandemic-induced strains on healthcare systems.  
- Measures like **Quietness of Hospital Environment** showed minimal change, indicating that certain aspects of patient experience were less impacted.  
- The consistent downward trend across most measures highlights the broad impact of the pandemic on healthcare service quality and patient satisfaction.  


```python

# Second Graph: Change in Overall Hospital Rating Before vs After COVID by State

# Define measure and year split
measure = 'Overall Hospital Rating'

# Calculate averages for each state before and after COVID-19
before_covid_state = hcahps_data[(hcahps_data['Year'] < year_split) & (hcahps_data['Measure'] == measure)] \
    .groupby('State Name')['Top-box Percentage'].mean().reset_index()
after_covid_state = hcahps_data[(hcahps_data['Year'] >= year_split) & (hcahps_data['Measure'] == measure)] \
    .groupby('State Name')['Top-box Percentage'].mean().reset_index()

# Merge and calculate the difference
before_after_covid_state = before_covid_state.merge(after_covid_state, on='State Name', suffixes=('_before', '_after'))
before_after_covid_state['Difference'] = before_after_covid_state['Top-box Percentage_after'] - before_after_covid_state['Top-box Percentage_before']
before_after_covid_state = before_after_covid_state.sort_values('Difference', ascending=False)

# Plot the second graph
national_avg_change = before_after_covid_state['Difference'].mean()
colors = ['#1f77b4' if diff >= 0 else '#d62728' for diff in before_after_covid_state['Difference']]
fig, ax = plt.subplots(figsize=(14, 10), dpi=100)

ax.barh(before_after_covid_state['State Name'], before_after_covid_state['Difference'], color=colors, edgecolor='black')

# Add average change line and label
ax.axvline(national_avg_change, color='black', linestyle='--', linewidth=1)
ax.text(-1.2, 10, f'National Avg Change ({national_avg_change:.2f})', color='black', fontsize=12, ha='right')

# Customize plot
ax.set_title('Change in Overall Hospital Rating Before vs After COVID by State', fontsize=16, weight='bold')
ax.set_xlabel('Difference in Score', fontsize=14)
ax.set_ylabel('State', fontsize=14)
ax.grid(True, axis='x', linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Add text labels for each bar
for i, (diff, state) in enumerate(zip(before_after_covid_state['Difference'], before_after_covid_state['State Name'])):
    ax.text(diff, i, f'{diff:.2f}', ha='right' if diff < 0 else 'left', va='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

## 5. Advanced Modeling & Statistical Analysis


<a id="51-modeling-objective--approach"></a>
### 5.1 Modeling Objective & Approach

- **Purpose**: Identify top drivers of *Overall Hospital Rating* (which measures matter most?).  
- **Models**: LassoCV, RidgeCV, ElasticNetCV with 5-fold cross-validation.  
- **Checks**: Residual plots to ensure no major violations (linearity, normality).


<a id="52-data-preparation--feature-engineering"></a>
### 5.2 Data Preparation & Feature Engineering


```python
# Filter relevant columns and create the 'covid_19' indicator
model_data = hcahps_data[['Year', 'State', 'State Name', 'Measure', 'Top-box Percentage']].copy()
model_data['covid_19'] = model_data['Year'].apply(lambda x: 1 if x in [2021, 2022] else 0)

# Pivot the data so each measure becomes a column
model_wide = model_data.pivot_table(
    index=['Year', 'State', 'State Name', 'covid_19'],
    columns='Measure',
    values='Top-box Percentage'
).reset_index()

# 3. Ensure the target variable exists and drop rows with missing target values
if 'Overall Hospital Rating' not in model_wide.columns:
    raise ValueError("The 'Overall Hospital Rating' column is missing.")
model_wide = model_wide.dropna(subset=['Overall Hospital Rating'])

# 4. Compute correlation and optionally drop a highly correlated feature
if 'Willingness to Recommend the Hospital' in model_wide.columns:
    correlation = model_wide['Overall Hospital Rating'].corr(
        model_wide['Willingness to Recommend the Hospital']
    )
    print("Correlation between 'Overall Hospital Rating' and 'Willingness to Recommend the Hospital':", correlation)
    model_wide = model_wide.drop(columns=['Willingness to Recommend the Hospital'])

# 5. Standardize column names
model_wide.columns = model_wide.columns.str.lower().str.replace(' ', '_')

# 6. Drop columns not required for modeling
transformed_data = model_wide.drop(columns=['year','state', 'state_name'])

# Display the transformed dataset
transformed_data.head()
```

#### 5.2.1 Define Target & Features

```python
# Define target and features
target = 'overall_hospital_rating'
features = [col for col in transformed_data.columns if col != target]

# Prepare X (features) and y (target)
X = transformed_data[features].copy()
y = transformed_data[target].copy()

# Remove rows with missing feature values and update target index
X = X.dropna(axis=0, how='any')
y = y.loc[X.index]

# Display the shape of X and y
print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)
```

#### 5.2.2 Train/Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,  # 20% for testing
    random_state=42
)
print("Train set:", X_train.shape, y_train.shape)
print("Test set:", X_test.shape, y_test.shape)
```

#### 5.2.3 Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler_cols = [c for c in features if c != 'covid_19']
scaler = StandardScaler()
X_train[scaler_cols] = scaler.fit_transform(X_train[scaler_cols])
X_test[scaler_cols] = scaler.transform(X_test[scaler_cols])
```

<a id="53-model-selection--cross-validation"></a>
### 5.3 Model Selection & Cross-Validation

```python
from sklearn.linear_model import LassoCV, RidgeCV, ElasticNetCV
from sklearn.metrics import mean_squared_error, r2_score

# Define a helper function to evaluate models
def evaluate_model(model, X_tr, y_tr, X_te, y_te, name=""):
    model.fit(X_tr, y_tr)
    pred_tr = model.predict(X_tr)
    pred_te = model.predict(X_te)

    # Calculate performance metrics
    rmse_tr = np.sqrt(mean_squared_error(y_tr, pred_tr))
    rmse_te = np.sqrt(mean_squared_error(y_te, pred_te))
    r2_tr = r2_score(y_tr, pred_tr)
    r2_te = r2_score(y_te, pred_te)

    # Display evaluation results
    print(f"\n=== {name} ===")
    print(f"Train RMSE: {rmse_tr:.3f}, R²: {r2_tr:.3f}")
    print(f"Test  RMSE: {rmse_te:.3f}, R²: {r2_te:.3f}")

# Lasso regression with cross-validation
lasso = LassoCV(cv=5, random_state=42)
evaluate_model(lasso, X_train, y_train, X_test, y_test, name="Lasso")

# Elastic Net regression with cross-validation
enet = ElasticNetCV(cv=5, random_state=42)
evaluate_model(enet, X_train, y_train, X_test, y_test, name="Elastic Net")

# Ridge regression with cross-validation
ridge = RidgeCV(cv=5)
evaluate_model(ridge, X_train, y_train, X_test, y_test, name="Ridge")

```

**Analysis:**
- All three models performed well with high R² values (above 0.93), indicating strong predictive power.
- The Lasso model had the best generalization, achieving the lowest test RMSE (0.995).
- Elastic Net and Ridge performed similarly but slightly underperformed compared to Lasso on test data.

<a id="54-residual-analysis--diagnostics"></a>
### 5.4 Residual Analysis & Diagnostics


```python
# Predict on the test set
y_test_pred = lasso.predict(X_test)

# Calculate residuals
residuals = y_test - y_test_pred

# 1. Residuals vs. Predicted values plot (Check for homoscedasticity)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_pred, y=residuals, color='blue')
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs. Predicted Values (Homoscedasticity Check)')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

# 2. Histogram of residuals (Check for normality)
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='purple')
plt.title('Histogram of Residuals (Normality Check)')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 3. QQ plot (Check for normality)
import scipy.stats as stats
import statsmodels.api as sm

plt.figure(figsize=(10, 6))
sm.qqplot(residuals, line='s')
plt.title('QQ Plot of Residuals')
plt.grid(True)
plt.show()
```

- No major heteroscedasticity.
- Residuals roughly normal => linear modeling assumptions hold.

<a id="55-feature-importance--interaction-analysis"></a>
### 5.5 Feature Importance & Interaction Analysis

```python
# Extract feature importance from the Lasso model
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': lasso.coef_
})

# Sort by the absolute value of coefficients
feature_importance['Abs_Coefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='Abs_Coefficient', ascending=False)

# Plot the feature importance
plt.figure(figsize=(12, 6))
sns.barplot(x='Coefficient', y='Feature', data=feature_importance, palette='viridis')
plt.title('Feature Importance from Lasso Model')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Display the feature importance table
feature_importance[['Feature', 'Coefficient']]

```

- **Care Transition**, **Quietness of Hospital Environment**, and **Cleanliness of Hospital Environment** are the most influential features with the highest positive coefficients.  
- **Communication with Doctors** has a negative but small impact, while **Covid_19** also shows a minor effect.  
- Features like **Communication about Medicines** and **Responsiveness of Hospital Staff** have moderate positive contributions to the prediction.  
- The Lasso model emphasizes key areas in hospital experience that strongly influence the overall rating.


#### 5.5.1 SHAP Summary for deeper insights
- SHAP analysis provides additional context by capturing case-specific variability and non-linear interactions, complementing the linear interpretation of the model.


```python
# Train a Gradient Boosting Regressor for SHAP analysis
from sklearn.ensemble import GradientBoostingRegressor
import shap

# Initialize and train the model
gbr = GradientBoostingRegressor(n_estimators=100, random_state=42)
gbr.fit(X_train, y_train)

# Predict on the test set
preds = gbr.predict(X_test)

# Evaluate the model's performance
rmse_gbr = np.sqrt(mean_squared_error(y_test, preds))
r2_gbr = r2_score(y_test, preds)
print(f"Gradient Boosting RMSE: {rmse_gbr:.3f}, R²: {r2_gbr:.3f}")

# SHAP analysis
explainer = shap.Explainer(gbr, X_train)
shap_values = explainer(X_test)

# Global feature importance plot
shap.summary_plot(shap_values, X_test, plot_type='bar', show=True)

# SHAP summary plot for all feature effects
shap.summary_plot(shap_values, X_test)

```

- **Care Transition**, **Quietness**, and **Cleanliness** exhibit high SHAP values, aligning with their strong positive coefficients in the Lasso model.  
- **Communication with Doctors** and **Communication about Medicines** have low or negative SHAP values, which also matches their small or negative coefficients.

<a id="6-findings--insights"></a>
## 6. Findings & Insights

1. **Pandemic Dip**: Overall Rating declined ~2–3% around 2021–2022.  
2. **Core Drivers**: *Care Transition*, *Quietness*, *Cleanliness*, and *Discharge Info* strongly elevate satisfaction.  
3. **Communication about Medicines**: Scores lag behind other measures, especially post-COVID.  
4. **Regional Variation**: Certain states improved despite pandemic pressures, while others had steeper declines.

<a id="7-recommendations--next-steps"></a>
## 7. Recommendations & Next Steps

### 7.1 Recommendations
- **Strengthen Care Transitions**: Thorough discharge protocols, follow-up calls, etc.  
- **Enhance Communication**: Medication education and staff responsiveness training.  
- **Maintain Clean & Quiet Environment**: Easier high-impact factor to control.  
- **Benchmarking**: Emulate best practices from states/hospitals that thrived.

### 7.2 Next Steps
- **Investigating States with Increased Positive Responses**: Understand why certain states reported improved patient responses during COVID-19, despite national trends showing a decline.
- **Correlation Analysis**: Explore potential relationships between patient satisfaction and state-level COVID-19 data, such as case counts and death rates, over time.
- **Data Integration**: Incorporate additional data (COVID-19 cases, staff-to-patient ratios, and hospital capacity) to enhance analysis accuracy.
-  **Interactive Dashboard**:Develop a dashboard for real-time performance monitoring, helping hospital leadership respond quickly to changes.
